8. What is backpropagation and how does it work? What is the difference between backpropagation and reverse-mode autodiff?
    - Backpropagation is Gradient Descent using an efficient technique for computing the gradients automatically (2 passes: forward, backward), and able to compute the gradient of the network's error with ragard to every single model parameter.
    - The **difference** between the two is that:
        - Backpropagation use reverse-mode autodiff to automatically calculate gradients -> Backpropagation is **the whole process**
        - Reverse-mode autodiff is just a **technique** to compute gradients efficiently
