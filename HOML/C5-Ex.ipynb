- Soft margin -> finding a compromise between perfectly seperating and. having widest possible street

2. What is a support vector?
- Supprt vectors are instances of points used to determine the boundary of each class (located of edge of the street)
- Answer:
    - Is instance located on the "street" (including border)
    - Decision boundary is entirely determined by these
    - Any instance which is not these has NO INFLUENCE -> can remove, move,.. as long as stay off the street
    - Computing the predection involves only these ones.
