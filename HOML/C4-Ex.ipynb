1. Which Linear Regression training algorithm can you use if you have a training set with millions of features?
- Batch/ Stochastic GD/ Mini-batch GD: since it doesn't have to deal with matrix inversion (which takes a lot of time for a huge features matrices)


2. Suppose the features in your training set have very different scales. Which algorithms might suffer from this, and how? What can you do about it?
- We should use Feature Scaling so that we can normalize all the feature
- All algorithm using Gradient Descent might suffer from this since it would makes the contour strechted, which might be hard to find the global minimum

3. Can Gradient Descent get stuck in a local minimum when training a Logistic Regression model?
- No. Because the cost function is convex, so you just have to have a good learning rate and some time)

4. Do all Gradient Descent algorithms lead to the same model, provided you let them run long enough?
- No, because of the different cost function

5.Suppose you use Batch Gradient Descent and you plot the validation error at every epoch. If you notice that the validation error consistently goes up, what is likely going on? How can you fix this?
- If validation error goes up, maybe learning rate is too high -> algorithm is diverging
- If trainig error also goes up -> clearly have to reduce learning rate
- If training error not goes up -> overfitting -> stop training to change the model

6. Is it a good idea to stop Mini-batch Gradient Descent immediately when the validation error goes up?
- No because Mini-batch will jump around alot so if we immediately stop, we might not be around the global minimum yet.

7. Which Gradient Descent algorithm (among those we discussed) will reach the vicinity of the optimal solution the fastest? Which will actually converge? How can you make the others converge as well?
- Stochastic Gradient Descent will reach the vicinity of the optimal solution the fastest
- Batch GD will actually converge
- Other will converge if you reduce the learning rate as we get into the core

8. Suppose you are using Polynomial Regression. You plot the learning curves and you notice that there is a large gap between the training error and the validation error. What is happening? What are three ways to solve this?
- Large gap means that the variance is high, which means it overfit the training set
- Solution:
    - Add more data
    - Regularize
    - Reduce polynomial degrees


9. Suppose you are using Ridge Regression and you notice that the training error and the validation error are almost equal and fairly high. Would you say that the model suffers from high bias or high variance? Should you increase the regularization hyperparameter  or reduce it?
- High bias, since it is underfitting
- We should reduce it because it would make the parameters bigger

10. Why would you want to use:
- Ridge Regression instead of plain Linear Regression (i.e., without any regularization)?
    - Because regularization helps the model to:
        - Fit data
        - Keep model weights as small as possible
- Lasso instead of Ridge Regression?
    - Lasso will do some kind of feature selection by eliminating the weights of the least important features
    - So we should use Lasso when suspect **only a few features are useful**
- Elastic Net instead of Lasso?
    - When we want to mix both Lasso and Ridge, with a scalar to control which one to out perform the other
    - Because Lasso may behave erratically when the # features >= number of training instances or several features are highly correlated
