1. For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

(a) The sample size n is extremely large, and the number of predictors p is small.
- Flexible > Inflexible
- Since flexible model usually leads to high variance, an extremely large sample size will be necessary for reducing it, which this is the case.

(b) The number of predictors p is extremely large, and the number of observations n is small.
- Flexible < Inflexible
- Due to small observations, flexible model can't high variance.

(c) The relationship between the predictors and response is highly non-linear.
- Flexible > Inflexible
- The non-linear relationship means that the model should be complex, which an inflexible model would lead to high bias.

(d) The variance of the error terms, i.e. σ2 = Var(ε), is extremely high.
- Flexible < Inflexible
- A flexible model might try to try to fit the noises, which leads to high variance
